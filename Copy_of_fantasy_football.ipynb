````python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
from numpy.random import default_rng
from time import sleep

#import geocoder

from geopy.geocoders import Nominatim

rng = default_rng()

def get_lat_long_geopy(location):
    """Gets latitude and longitude for a given location using GeoPy.

    Args:
        location (str): The location string (e.g., "Tyler, TX").

    Returns:
        tuple: A tuple containing latitude and longitude (float, float) if successful,
               or (None, None) if there's an error or no results.
    """
    try:
        geolocator = Nominatim(user_agent="fantasy_football_birthplace_lookup")
        g = geolocator.geocode(location, featuretype='city')
        #g = geocoder.osm(location)
        if g:
          return g
        else:
            print(f"GeoPy couldn't find coordinates for location: {location}")
            return None, None
    except Exception as e:
        print(f"Error using GeoPy: {e}")
        return None, None

def birthdate(pfr_id="DextGe00"):
  if pfr_id == 'NaN':
    location = None
    latitude = None
    longitude = None
    return location, latitude, longitude
  time_to_sleep = rng.uniform(5, 8)
  sleep(time_to_sleep)
  print(pfr_id)
  url = f"https://www.pro-football-reference.com/players/{pfr_id[0].upper()}/{pfr_id}.htm"
  response = requests.get(url)
  response.raise_for_status()

  soup = BeautifulSoup(response.content, 'html.parser')

  birth_span = soup.find(id='necro-birth')

  if birth_span:
    birth_date = birth_span.get('data-birth')

    location_match = re.search(r'in\s(.*?)<a href="/friv/birthplaces.*?>(.*?)</a>', str(birth_span))
    if location_match:
        city = location_match.group(1).strip()
        state = location_match.group(2).strip()
        location = f"{city} {state}"
    else:
        location = None
        latitude = None
        longitude = None
    '''
    if location:
      geoloc = get_lat_long_geopy(location)
      if geoloc:
        latitude, longitude = geoloc.latitude, geoloc.longitude
      else:
        latitude = None
        longitude = None
    '''
    #data = {'birth_date': [birth_date], 'birth_place': [location], 'latitude': [latitude], 'longitude': [longitude]}
    #df = pd.DataFrame(data)
    return birth_date, location #, latitude, longitude
  else:
    print("Error: Could not find the element with id='necro-birth' on the page.")
    location = None
    birth_date = None
    return birth_date, location

print(birthdate('LaPoSa01'))
````

````python
from kerykeion import AstrologicalSubject, KerykeionChartSVG
import base64

def astro(name, birthdate, birthtime, team, number, position, bt, lat, lng, tz_str, chart=False, online=False, debug=False):
  print(name)
  planets = ['sun',
             'moon',
             'mercury',
             'venus',
             'mars',
             'jupiter',
             'saturn',
             'uranus',
             'neptune',
             'pluto',
             'true_node',
             'mean_node',
             'chiron',
             'mean_lilith']

  houses = ['first_house',
            'second_house',
            'third_house',
            'fourth_house',
            'fifth_house',
            'sixth_house',
            'seventh_house',
            'eighth_house',
            'ninth_house',
            'tenth_house',
            'eleventh_house',
            'twelfth_house']

  k = planets + houses
  y, m, d = birthdate.timetuple()[:3]
  if bt:
    bt = '⌖'
  else:
    bt = '◌'
  hr, mn = birthtime.split(":")[:2]
  hr, mn = int(hr), int(mn)
  a = AstrologicalSubject(name, y, m, d, hr, mn, city=f'{team}-{number}-{position} {bt}', lat=lat, lng=lng, nation='NFL', tz_str=tz_str, online=False)
  astro_dict = dict
  if debug == True:
    return a
  if chart == False:
    return dict((key, dict(a[key])) for key in k)
  if chart == True:
    c = base64.b64encode(KerykeionChartSVG(a).makeTemplate(minify=True).encode('utf-8'))
    _ = dict((key, dict(a[key])) for key in k)
    _['chart'] = c
    return _
````
